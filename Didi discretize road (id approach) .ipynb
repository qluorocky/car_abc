```python
def pop_path():
    l = 0.01
    while pathes:
        _, path = pathes.popitem()
        if (len(path) >= 10 
            and 3.067947e+01 -l < path[0,2] < 3.067947e+01 + l
            and 3.067947e+01 -l < path[-1,2] < 3.067947e+01 + l
            and  1.040783e+02 - l < path[0,1] < 1.040783e+02 + l
            and  1.040783e+02 - l < path[-1,1] < 1.040783e+02 + l):
            break
    return path
```

```python
def build_table(roads_small_id):
    s = set()
    for i in roads_small_id:
        s |= set(i)
    table = dict((k, {}) for k in s)
    for path in roads_small_id:
        for i in range(len(path) - 1):
            j = i + 1
            p1,p2 = path[i], path[j]
            if p2 not in table[p1]:
                table[p1][p2] = 1
            else:
                table[p1][p2] += 1
    return table
```
